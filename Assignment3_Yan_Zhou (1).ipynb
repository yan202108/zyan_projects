# **Part2 Text Preprocessing**

**Change the raw text to lower case**

```python
alice_raw = alice_raw.lower()
```

```python
alice_raw
```

**Tokenize the sentences**

```python
import nltk
nltk.download('punkt')
from nltk import tokenize
```

```python
txt_sent=tokenize.sent_tokenize(alice_raw)
```

```python

type(txt_sent), len(txt_sent)
```

```python
txt_words = [tokenize.word_tokenize(sent) for sent in txt_sent]
print(txt_words[:2])
```

```python
sum_word=0
for i in range(len(txt_sent)):
    sum_word = sum_word+len(txt_words[i])
print(sum_word)
```

**Import punctuation from the string module** 

```python
from string import punctuation
```

```python
list_punct=list(punctuation)
```

```python
len(list_punct)
```

```python
def drop_punct(input_tokens):
    return [token for token in input_tokens if token not in list_punct]
```

```python

txt_words_nopunct = [drop_punct(sent) for sent in txt_words]
print(txt_words_nopunct)
```

```python
sum_word_np=0
for i in range(len(txt_words_nopunct)):
    sum_word_np = sum_word_np+len(txt_words_nopunct[i])
print(sum_word_np)
```

```python
abs(sum_word_np-sum_word)
```

**Import the stop words from NLTK**

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
list_stop=stopwords.words('english')
print(len(list_stop))
```

**Create a variable holding the contextual stop words**

```python
cont_stop=['said','--']
```

**Create a master list concatenate punctuation, stop words and contexual stop words**

```python
m_list=list_punct+list_stop+cont_stop
```

```python
print(len(m_list))
```

**Define a funtion to drop these tokens from any input sentence**

```python
def drop_com(input_tokens):
    return [token for token in input_tokens if token not in m_list]
```

```python
txt_words_clean = [drop_com(sent) for sent in txt_words]
print(txt_words_clean)
```

```python
sum_word_clean=0
for i in range(len(txt_words_clean)):
    sum_word_clean = sum_word_clean+len(txt_words_clean[i])
print(sum_word_clean)
```

**Stemming**

```python
from nltk.stem import PorterStemmer
```

```python
stemmer_p=PorterStemmer()
```

```python
def stem_word(input_word):
    return [stemmer_p.stem(word) for word in input_word]
```

**Print the first 5 sentences from the result**

```python
txt_clean_stem=[stem_word(word) for word in txt_words_clean]
for i in range(5):
    print("sentence",(i+1),": ", txt_clean_stem[i],'\n')
print ('Compare with the sentences without stemming:','\n')
for i in range(5):
    print("sentence",(i+1),": ", txt_words_clean[i],'\n')
```

# **Part3 Text Preprocessing**

**Import word2vec and train the word embeddings**

```python
alice_words_nonstop=txt_words_clean
print(alice_words_nonstop[:3])
```

```python
import gensim.downloader as api
from gensim.models import word2vec
import numpy as np
```

```python
np.random.seed(1)
```

```python
model = word2vec.Word2Vec(alice_words_nonstop)
```

```python
print(model.wv["rabbit"])
```

**Find the term most similar to "rabbit"**

```python
model.wv.most_similar("rabbit")
```

**using window size 2, retrain the word vectors, and find the term most similar to "rabbit"**





```python
model_w2 = word2vec.Word2Vec(alice_words_nonstop,size=2)
```

```python
model_w2.wv.most_similar("rabbit")
```

**using Skip-gram method with window size 5, retrain the word vectors, and find the term most similar to "rabbit"**

```python
model_sg = word2vec.Word2Vec(alice_words_nonstop, sg=5)
```

```python
model_sg.wv.most_similar("rabbit")
```

**Find the representattion for the phrase "white rabbit" by averaging the vectors for "white" and "rabbit"**

```python
v1 = model.wv['white']
v2 = model.wv['rabbit']
res1 = (v1+v2)/2
```

```python
print(res1)
```

**Find the representattion for the phrase "mad hatter" by averaging the vectors for "white" and "rabbit"**

```python
v3 = model.wv['mad']
v4 = model.wv['hatter']
res2 = (v3+v4)/2
```

```python
print(res2)
```

**Find the cosine similarity between these two phrases**

```python
model.wv.cosine_similarities(res1, [res2])
```
